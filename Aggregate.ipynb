```python
import pandas as pd
import numpy as np
df = pd.read_csv('bigmart_data.csv')
```

```python
price_by_type = df.groupby('Item_Type')
price_by_type.first()
```

```python
price_by_type.Item_MRP.mean()
```

```python
price_by_type_n_fat = df.groupby(['Item_Type','Item_Fat_Content'])
price_by_type_n_fat.Item_MRP.mean()
```

Crosstab

```python
pd.crosstab(df['Item_Type'],df['Item_Fat_Content'])
```

Pivot Table

```python
pd.pivot_table(df,index=['Item_Type','Item_Fat_Content'],values=['Item_Outlet_Sales'],aggfunc=[sum,np.median,np.mean])
```
