```python
import pandas as pd
```

```python
df = pd.read_csv('bigmart_data.csv')
df.head()
```

Sorting

```python
df.sort_values('Item_Weight').head(10)
```

Sort Descending

```python
df.sort_values('Item_Weight',ascending=False).head(10)
```

Reflect Sort in actual data fram by using inplace arguement

```python
df_inplace = pd.read_csv('bigmart_data.csv')
df_inplace.sort_values('Item_MRP', inplace=True)
df_inplace.head()
```

Sorting on multiple columns: Order precedence of columns

```python
df.sort_values(by=['Outlet_Establishment_Year','Item_Outlet_Sales'],ascending=[False,True], inplace=True)
df.head()
```

Sort by index order

```python
df.sort_index()
```
