Introduction to Matplotlib library

```python
import matplotlib.pyplot as plt
import pandas as pd, numpy as np
```

Basic plot

```python
height = [150,160,165,185]
weight = [70,80,90,100]

plt.plot(height,weight)
```

Customizing plot

```python
plt.title('Height-Weight Relationship')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.plot(height,weight)
```

Multiple lines in one plot

```python
calories_burnt = [65,75,95,99]
plt.plot(calories_burnt)
plt.plot(weight)
```

Adding legend for multiple lines

```python
plt.plot(calories_burnt)
plt.plot(weight)
plt.legend(labels=['Calories Burnt','Weight'], loc='lower right')
```

Labeled Values / Ticks

```python
plt.plot(calories_burnt)
plt.plot(weight)
plt.legend(labels=['Calories Burnt','Weight'], loc='lower right')
plt.xticks(ticks=[0,1,2,3], labels=['p1','p2','p3','p4'])
```

Customizing figure

```python
plt.figure(figsize=(15,5))
plt.plot(calories_burnt)
plt.plot(weight)
plt.legend(labels=['Calories Burnt','Weight'], loc='lower right')
plt.xticks(ticks=[0,1,2,3], labels=['p1','p2','p3','p4'])
```

Customizing colors and shape of plot. Default 'b-' i.e. Blue Slid line. 'go' Green Dots, 'y--' Yellow dotted line.

```python
plt.plot(calories_burnt,'go')
plt.plot(weight,'y--')
plt.legend(labels=['Calories Burnt','Weight'], loc='lower right')
plt.xticks(ticks=[0,1,2,3], labels=['p1','p2','p3','p4'])
```

Subplots

```python
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(6,3),sharey=True)
ax[0].set_title('Calories Burnt')
ax[0].plot(calories_burnt,'go')
ax[0].set_xticks(ticks=[0,1,2,3])
ax[0].set_xticklabels(labels=['p1','p2','p3','p4'])

ax[1].set_title('Weight')
ax[1].plot(weight,'y--')
ax[1].set_xticks(ticks=[0,1,2,3])
ax[1].set_xticklabels(labels=['p1','p2','p3','p4'])
```
